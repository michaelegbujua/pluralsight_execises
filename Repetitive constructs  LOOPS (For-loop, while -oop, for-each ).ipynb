# WHILE LOOP

A While loop is used when the user do not the specific number of times the program will run , base on a certain logical or arithmetic condition evaluates to true

```python
while False:
    print(' will be printed when the condition evaluates to true')
```

```python
while True:
    print('will be printed when the condition evaluates to true ')
```

the program will run infinetely since it continues to evaluate as true

```python
a = 5
while a < 10:
    print(a)
```

when working with while loop you must specify the conditions correctly in order to avoid entering into infinite loops

```python
b = 7
while b <= 7:
        print(b)
        b = b * 1
```

```python
b
```

```python
c = 45
while c < 50:
    print(c)
    c = c + 1
```

```python
c
```

```python
 d = 64
while d <= 75:
        print(d)
        d += 3
```

```python
p = 5
while p:
    print('P is still Positive:', p)
    p -= 1
```
